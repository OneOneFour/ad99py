# Coarsen Loon

```python
import numpy as np
import pandas as pd

df = pd.read_csv('data/loon/non_orographic_flights.csv')
```

```python
df['time'] = pd.to_datetime(df.time,format='ISO8601')
```

```python
coarsen_track = df.groupby('flight_id').resample('d',on='time').sum()
```

```python
coarsen_track
```

```python
df[df['flight_id'] == 'L-002'].resample('h',on='time').sum()
```

```python

flights = []
for flight in df['flight_id'].unique():
    ## this is bad code, but it works
    flight_df = df[df['flight_id'] == flight]
    flight_df = flight_df.resample('D',on='time').agg({
    'latitude': 'mean',     # sum this column
    'longitude': 'mean',    # average this column
    'flux_east': ['sum','mean'],
    'flux_west': ['sum','mean'],
    'flux_north':['sum','mean'],
    'flux_south':['sum','mean'],    # take first value
    })
    flight_df.columns = [
        f"{col[0]}_{col[1]}" if isinstance(col, tuple) else col
        for col in flight_df.columns
    ]
    flight_df['flight_id'] = flight
    flight_df = flight_df.set_index([flight_df.index,'flight_id'])

    flights.append(flight_df)
coarsen_flights = pd.concat(flights)
coarsen_flights = coarsen_flights.dropna()
```

```python
coarsen_flights
```

```python
coarsen_flights.to_csv('data/loon/coarsen_flights_nonoro_daily.csv')
```

```python
eastern_flux_coarse = coarsen_flights['flux_east_sum'].to_numpy()*1000
eastern_flux_coarse_mean = coarsen_flights['flux_east_mean'].to_numpy()*1000
eastern_flux = df['flux_east'].to_numpy()*1000

western_flux_coarse = coarsen_flights['flux_west_sum'].to_numpy()*1000
western_flux = df['flux_west'].to_numpy()*1000

net_zonal_flux_coarse = eastern_flux_coarse + western_flux_coarse
abs_zonal_flux_coarse = np.abs(western_flux_coarse) + np.abs(eastern_flux_coarse)

net_zonal_flux = eastern_flux + western_flux
abs_zonal_flux = np.abs(western_flux) + np.abs(eastern_flux)


eastern_flux_coarse = eastern_flux_coarse[eastern_flux_coarse > 0]
eastern_flux = eastern_flux[eastern_flux > 0]   
eastern_flux_coarse_mean = eastern_flux_coarse_mean[eastern_flux_coarse_mean > 0]   
net_zonal_flux_coarse = net_zonal_flux_coarse[net_zonal_flux_coarse > 0]
abs_zonal_flux_coarse = abs_zonal_flux_coarse[abs_zonal_flux_coarse > 0]
net_zonal_flux = net_zonal_flux[net_zonal_flux > 0]
abs_zonal_flux = abs_zonal_flux[abs_zonal_flux > 0]
```

```python
import matplotlib.pyplot as plt 
import numpy as np



native_hist,native_edges = np.histogram(np.log10(eastern_flux),bins=100,density=True)
native_centers = (native_edges[:-1] + native_edges[1:])/2
coarsen_hist,coarsen_edges = np.histogram(np.log10(eastern_flux_coarse),bins=100,density=True)
coarsen_centers = (coarsen_edges[:-1] + coarsen_edges[1:])/2

coarsen_mean_hist,coarsen_mean_edges = np.histogram(np.log10(eastern_flux_coarse_mean),bins=100,density=True)
coarsen_mean_centers = (coarsen_mean_edges[:-1] + coarsen_mean_edges[1:])/2

plt.plot(native_centers,native_hist, label='2 minute resolution')
plt.plot(coarsen_centers,coarsen_hist, label='1 hourly cumulative')
plt.plot(coarsen_mean_centers,coarsen_mean_hist, label='1 hourly mean')
plt.legend()
plt.xlim(-3,3)
plt.ylabel("PDF")
plt.xlabel("log10(GW Eastward Flux) [mPa]")
```

```python
native_total, native_edges = np.histogram(np.log10(net_zonal_flux),bins=100,density=True)
native_centers = (native_edges[:-1] + native_edges[1:])/2
coarsen_total, coarsen_edges = np.histogram(np.log10(net_zonal_flux_coarse),bins=100,density=True)
coarsen_centers = (coarsen_edges[:-1] + coarsen_edges[1:])/2

plt.plot(native_centers,native_total, label='2 minute resolution')
plt.plot(coarsen_centers,coarsen_total, label='1 hourly cumulative')
plt.legend()
plt.xlabel("log10(GW Net Zonal Flux) [mPa]")
plt.ylabel("PDF")
```
